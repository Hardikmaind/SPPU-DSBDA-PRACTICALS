```python
x = df.iloc[:, [2,3]].values
```

```python
x
```

```python
#The code x = df.iloc[:, [2,3]].values extracts a subset of columns from a pandas DataFrame df and assigns it to the variable x. Specifically, it selects the columns at index positions 2 and 3 (0-based indexing) and converts them into a NumPy array using the .values attribute.


#Here's a breakdown of the code:

#df: The pandas DataFrame from which the subset of columns will be extracted.
#iloc: The .iloc indexer in pandas is used to access DataFrame elements by integer-based indexing.
#[:, [2,3]]: The first : selects all rows, and [2,3] specifies the column indices to be included. This means that columns at index positions 2 and 3 will be selected.
#.values: The .values attribute converts the selected DataFrame subset into a NumPy array.

#After executing this code, the variable x will contain a NumPy array consisting of the values from the selected columns of the DataFrame. Each row in x corresponds to a row in the original DataFrame, and the selected columns are included in the same order as they appear in the DataFrame.
```

```python
y = df.iloc[:, 4].values
```

```python
y
```

```python
#splitting Data into 75% training and 25% testing data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=3)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test
```

```python
x_test.shape
```

```python
y_test
```

```python
y_test.shape
```

```python
#Data Standardization
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Initializing logistic Regression model and training the model
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state= 0)
classifier.fit(x_train, y_train)
```

```python
#Predction
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#Confusion Matrix


#A confusion matrix is a table that summarizes the performance of a 
#classification model by displaying the counts of true positive, true negative, 
#false positive, and false negative predictions. It is a useful tool for 
#evaluating the performance of a classification algorithm and understanding 
#the types of errors it makes.
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#Calculating accuracy of model
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
accuracy
```

```python
#True Positive: Actually True, Predicted True
```

```python
tp = cm[0,[0]]
print('True Positive : ', tp)
```

```python
#False Positive: Actually false, Predicted True
```

```python
fp = cm[0,[1]]
print('False Positive: ', fp)
```

```python
#False Negative : Actually true, Predicted False
```

```python
fn = cm[1, [0]]
print('False Negative: ', fn)
```

```python
#True Negative: Actually false, Predicted false
```

```python
tn = cm[1, [1]]
print('True Negative : ', tn)
```

```python
#Accuracy
```

```python
accuracy_cm = ((tp+tn)/(tp+fp+fn+tn))
print('Accuracy: ', accuracy_cm*100)
```

```python
#Error Rate
```

```python
error_rate_cm = ((fp+fn)/(tp+ fn+ fp+ tn))
print('Error Rate : ', error_rate_cm*100)
```

```python
#precision
```

```python
precision_cm = (tp/(tp+fp))
print('Precision : ', precision_cm*100)
```

```python
#Recall   recall is also called sensitivity
```

```python
recall_cm = (tp/(tp+fn))
print('Sensitivity : ', recall_cm*100)
```

```python
#Specificity
```

```python
specificity_cm = (tn/ (tn+fp))
print('Specificity : ', specificity_cm*100)
```
