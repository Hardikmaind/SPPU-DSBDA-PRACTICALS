```python
x = df.iloc[:, 0:4].values
x
```

```python
y = df.iloc[:,4].values
y
```

```python
#Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
#The LabelEncoder is a utility class in scikit-learn that can be used to encode categorical labels into numeric labels. It assigns a unique integer to each unique category in the target variable y.
#After fitting the LabelEncoder to the target variable using the fit_transform() method, the original categorical labels in y are replaced with the corresponding numeric labels.
#By performing this encoding, you can use the transformed y as the target variable in machine learning algorithms that require numeric inputs.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
#The train_test_split() function randomly shuffles and splits the input features x and the target variable y into two sets: the training set and the testing set. The test_size parameter specifies the proportion of the data that should be allocated to the testing set, in this case, 30% (0.3). The random_state parameter is used to ensure reproducibility by setting a specific random seed.
#After calling train_test_split(), the resulting splits are assigned to the variables x_train, x_test, y_train, and y_test. These variables will hold the training and testing sets for the input features and the target variable, respectively.
#You can use these splits to train your machine learning model on the training set (x_train and y_train) and evaluate its performance on the testing set (x_test and y_test).
```

```python
#Initializing Naive Bayes Model and training it
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#First, you import the GaussianNB class from the sklearn.naive_bayes module. This class represents the Gaussian Naive Bayes classifier, which is suitable for classification tasks when the input features are continuous.

#Next, you create an instance of the GaussianNB classifier using model = GaussianNB(). This creates an empty model object with default parameters.
#Finally, you fit the model to the training data using the fit() method, passing in x_train (the input features) and y_train (the target variable). This trains the model on the training data, allowing it to learn the underlying patterns and relationships between the input features and the target variable.
#Once the model is fitted, you can use it to make predictions on new, unseen data.
```

```python
#Prediction
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
y_test
```

```python
#Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction)
cm
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, prediction)*100
accuracy
```
